# Funciones, np.arrays y gr√°ficos 

```python
def funcionpartida(x):
    if x < 1:
        y = x+1
    if x >=1:
        y = 4*x-2
    return y
```

```python
x = np.linspace(0,2,100,endpoint=True)
```

```python
y = np.array([funcionpartida(t) for t in x])
```

```python
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.grid('on')
plt.title('Funci√≥n partida')
plt.savefig('Ejemplo1.png')
```

```python
def funcionpartidamejor(x):
    return np.where(x < 1, x+1, 4*x-2)
```

```python
x = 0.4
```

```python
funcionpartidamejor(x)
```

```python
newx = np.linspace(0,2,100)
```

```python
y = funcionpartidamejor(newx)
```

```python
plt.plot(newx, y)
```

# Escritura de archivos 

```python
newdata = 'ejemploescritura.dat'
```

```python
XY = np.vstack((newx, y))
```

```python
XY = XY.transpose()
```

```python
np.savetxt(newdata, XY, fmt='%5.3f', delimiter='\t\t', header='X  Y')
```

##  escritura secuencial

```python
with open('escriturasecuencial.txt', 'w') as f:
    f.write('# X Y\n')
    for thisx, thisy in zip(newx, y):
        f.write(f'{thisx:5.3f}  {thisy:5.3}\n')
    
```
